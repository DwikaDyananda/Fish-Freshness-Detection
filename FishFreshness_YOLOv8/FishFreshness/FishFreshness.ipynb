```python
def process_image(model, image_path, confidence_threshold):
    image = cv2.imread(image_path)
    results = model(image)

    detections = []
    for r in results:
        boxes = r.boxes
        for box in boxes:
            confidence = box.conf[0].item()
            if confidence >= confidence_threshold:
                x1, y1, x2, y2 = box.xyxy[0].tolist()
                class_id = box.cls[0].item()
                class_name = model.names[int(class_id)]

                # Attempt to determine part and freshness from class_name
                if '_' in class_name:
                    part, freshness = class_name.split('_', 1)
                else:
                    part = "unknown"
                    freshness = class_name

                detection = {
                    'part': part.lower(),
                    'label': freshness,
                    'confidence': confidence,
                    'bbox': [x1, y1, x2 - x1, y2 - y1]
                }
                detections.append(detection)

    # If parts are unknown, assume first detection is eye, second is gills
    if all(det['part'] == 'unknown' for det in detections):
        if len(detections) >= 1:
            detections[0]['part'] = 'mata'
        if len(detections) >= 2:
            detections[1]['part'] = 'insang'

    return detections

# Load your YOLO model
model = YOLO('/content/best_2.pt')

# Process the image with a confidence threshold
image_path = '/content/ikan_full.jpg'
confidence_threshold = 0.25
detections = process_image(model, image_path, confidence_threshold)

# Freshness determination function
def determine_fish_freshness(detections):
    def evaluate_part(detection):
        label, confidence = detection['label'], detection['confidence']
        if label == "Fresh" and confidence > 0.30:
            return True
        return False

    eye_fresh = gills_fresh = None
    for detection in detections:
        if detection['part'] == 'mata':
            eye_fresh = evaluate_part(detection)
        elif detection['part'] == 'insang':
            gills_fresh = evaluate_part(detection)

    if eye_fresh is None or gills_fresh is None:
      if eye_fresh is False or  gills_fresh is False:
        return "Non-Fresh"
      else:
        return "Insufficient data"
    return "Fresh" if eye_fresh and gills_fresh else "Non-Fresh"

# Determine overall freshness
result = determine_fish_freshness(detections)
print(f"Ikan termasuk Kategori: {result}")

# Print out all detections for verification
for det in detections:
    print(f"Part: {det['part']}, Label: {det['label']}, Confidence: {det['confidence']:.2f}, BBox: {det['bbox']}")
```
